### Red neuronal

Son modelos computacionales inspirados en el comportamiento de una neurona humana con dendritas, cuerpo, axón y sinapsis. En una neurona artificial, las dendritas representan las entradas de datos, el cuerpo (o soma) representa la neurona, el axón la función de activación (activa las señales que ingresan a la neurona) y la salida de los datos y la sinapsis el punto de interconexión con otras neuronas (en este caso empleamos los pesos sinápticos).

### Esquema de Red Neuronal

### Representación matemática de la transformación de los datos de entrada - Función Neta o z.

$
z = Neta_j = \sum_i x_i*w_{ji} + \theta_j, z \in I\!R
$

$
z = w^t X_j + \theta_j
$

$
x_{i} =
\begin{bmatrix}
x_{1} & x_{2} & x_{3}
\end{bmatrix}
$


$
x_j =
\begin{bmatrix}
x_1 \\ 
x_2 \\ 
x_3
\end{bmatrix}
$

$
w_{i,j} =
\begin{bmatrix}
w_{1,1} \\ 
w_{2,1} \\ 
w_{3,1}
\end{bmatrix}
$

$
w_{j,i} = w^t=
\begin{bmatrix}
w_{1,1} & w_{1,2} & w_{1,3}
\end{bmatrix}
$

Cada $i$ representa una entrada y cada $j$ representa una neurona. $\theta$ representa el sesgo, intercepto o bias. Cada $x_i$ representa una variable, para este caso se represento el primer dato de cada variable.

$
w^t * x_j=
\begin{bmatrix}
w_{1,1} & w_{1,2} & w_{1,3}
\end{bmatrix} *
\begin{bmatrix}
x_1 \\ 
x_2 \\ 
x_3
\end{bmatrix} = w_{1,1}*x_1 + w_{1,2}*x_2 + w_{1,3}*x_3 = \sum_{j=1}^n w_{j,i}*x_j
$


$
\phi_j=
\begin{bmatrix}
\phi_1
\end{bmatrix}
$

$
w^t * x_j=
\begin{bmatrix}
w_{1,1} & w_{1,2} & w_{1,3}
\end{bmatrix} *
\begin{bmatrix}
x_1 \\ 
x_2 \\ 
x_3
\end{bmatrix} + \phi_j
$

## Funciones de activación

Las funciones de activación se encargan de realizar la transformación no lineal de la entrada. Sin las funciones de activación una red neuronal solo sería un modelo de regresión lineal.

```python
x = np.arange(-1,1,0.01)
```

### Lineal

$
f(x)=x
$

```python
plt.plot(x,x)
plt.ylabel("f(x)")
plt.xlabel("x")
```

### De paso binario

$
f(x)=1, si \hspace{1mm} x \hspace{1mm} es >= 0
$

$
f(x)=0, si \hspace{1mm} x \hspace{1mm} es < 0
$

```python
y = []
for i in range(len(x)):
    if x[i]>=0:
        y.append(1)
    else:
        y.append(0)
plt.plot(x,y)
plt.ylabel("f(x)")
plt.xlabel("x")
```

### De paso bipolar

$
f(x)=1, si \hspace{1mm} x \hspace{1mm} es >= 0
$

$
f(x)=-1, si \hspace{1mm} x \hspace{1mm} es < 0
$

```python
y = []
for i in range(len(x)):
    if x[i]>=0:
        y.append(1)
    else:
        y.append(-1)
plt.plot(x,y)
plt.ylabel("f(x)")
plt.xlabel("x")
```

### Sigmoida binaria

$
f(x)= \frac{1}{1+e^{-\sigma x}}
$

$
\sigma = 1
$

```python
sigma=np.linspace(1,5,5)
for i in range(len(sigma)):

    y = 1/(1+np.exp(-sigma[i]*x))

    plt.plot(x,y, label=sigma[i])
    plt.ylabel("f(x)")
    plt.xlabel("x")
    plt.legend(title="$\sigma$")
```

### Sigmoida bipolar

$
g(x) = 2f(x) -1= \frac{1-e^{-\sigma x}}{1+e^{-\sigma x}}
$

```python
sigma=np.linspace(1,5,5)
for i in range(len(sigma)):

    y = (1-np.exp(-sigma[i]*x))/(1+np.exp(-sigma[i]*x))

    plt.plot(x,y, label=sigma[i])
    plt.ylabel("f(x)")
    plt.xlabel("x")
    plt.legend(title="$\sigma$")
```

### Tangente hiperbólica

$
f(x)=tanh(x)= \frac{2}{1+e^{-2x}}-1
$

```python
y = 2/(1+np.exp(-2*x)) - 1
plt.plot(x,y)
plt.ylabel("f(x)")
plt.xlabel("x")
```

### ReLu

$
f(x)= ReLu(x)= max(0,x)
$

```python
y = []
for i in range(len(x)):
    if x[i]>=0:
        y.append(x[i])
    else:
        y.append(0)
plt.plot(x,y)
plt.ylabel("f(x)")
plt.xlabel("x")
```

#### Ejemplo:

```python
np.random.seed(0)
x_i = np.random.rand(10)
print(x_i)
```

```python
yj= 10
```

```python
#fAct:Lin,PasBin,PasBid, SigBin, SigBid, tanh, ReLu
def RNA(x,neuronas,fAct):
    xj= np.transpose(x)
    wij = np.random.rand(len(x),neuronas)
    wji = np.transpose(wij)
    biasi = np.random.rand(neuronas)
    biasj = np.transpose(biasi)
    zj = np.dot(wji,xj)+biasj
    y_p = []
    if fAct=="Lin":
        fx = zj
        y_p.append(fx)
    elif fAct=="PasBin":
        for i in range(len(zj)):
            if zj[i]>=0:
                fx=1
                y_p.append(fx)
            else:
                fx=0
                y_p.append(fx)
    elif fAct=="PasBid":
        for i in range(len(zj)):
            if zj[i]>=0:
                fx=1
                y_p.append(fx)
            else:
                fx=-1
                y_p.append(fx)
    elif fAct=="SigBin":
        fx = 1/(1+np.exp(-1*zj))
        y_p.append(fx)
    elif fAct=="SigBip":
        fx = (1-np.exp(-1*zj))/(1+np.exp(-1*zj))
        y_p.append(fx)
    elif fAct=="ReLu":
        for i in range(len(zj)):
            if zj[i]>=0:
                fx=zj[i]
                y_p.append(fx)
            else:
                fx=0
                y_p.append(fx)
    else:
        fx = 2/(1+np.exp(-2*zj)) - 1
        y_p.append(fx)
    return np.array(y_p)
```

```python
yp1 = RNA(x_i,6, "ReLu")
yp1
```

```python
yp2 = RNA(yp1.flatten(),1, "ReLu")
yp2
```

### Aprendizaje de la red

El aprendizaje para una red neuronal artificial, es la habilidad para aprender del entorno y mejorar el desempeño, es un proceso iterativo que permite ajustar los pesos sinápticos.

$
w(t+1)= w(t)+ \Delta w(t)
$

$w(t+1)$: valor actualizado del peso sináptico

$w(t)$: valor actual del peso sináptico

$\Delta w(t)$: variación del peso sináptico

El índice de desempeño de la red ($J$), lo definimos a través del error cuadrático promedio para $P$ patrones.

$
J =\frac{1}{2P}\sum_{P=1}^{P} (y_{real}-y_{modelado})^2
$

Lo ideal es que el valor de $J$ sea mínimo, para ello hacemos proporcional la variación de pesos al gradiente negativo ($\alpha$).

$
\Delta w_i = -\alpha\frac{\partial J}{\partial w_i} = -\alpha * (y_{real}-y_{modelado})*(-1)*x_i
$

Entonces:

$
w(t+1)= w(t)+ \Delta w(t) = w(t) + \alpha * (y_{real}-y_{modelado})*x_i
$
